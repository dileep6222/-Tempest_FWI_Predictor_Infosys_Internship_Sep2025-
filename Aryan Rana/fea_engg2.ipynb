```python
data['FFMC__TEMP'] = data['temperature'] * data['fine_fuel_moisture_code_transformed']
data['FFMC_RH'] = data['relative_humidity'] * data['fine_fuel_moisture_code_transformed']
data['RH_TEMP'] = data['temperature'] * data['relative_humidity']
data['RD_FFMC'] =  data['fine_fuel_moisture_code_transformed'] * data['rain_days']
data['DC_RD'] =  data['drought_code_transformed'] * data['rain_days']
data['WS_TENP'] =  data['wind_speed_transformed'] * data['fine_fuel_moisture_code_transformed']
data['IST_WS'] =  data['wind_speed_transformed'] * data['initial_spread_index_transformed']
data['ISI_FFMC'] =  data['initial_spread_index_transformed'] * data['fine_fuel_moisture_code_transformed']
data['ISI_RH'] =  data['initial_spread_index_transformed'] * data['relative_humidity']
data['DMC_TEMP'] =  data['duff_moisture_code_transformed'] * data['temperature']
data['DMC_RD'] =  data['duff_moisture_code_transformed'] * data['rain_days']
data['RD_TEMP'] =  data['rain_days'] * data['temperature']
data['DMC_DC'] =  data['drought_code_transformed'] * data['duff_moisture_code_transformed']
data['BUI_DC'] =  data['drought_code_transformed'] * data['build_up_index_transformed']
data['BUI_DMC'] =  data['duff_moisture_code_transformed'] * data['build_up_index_transformed']
```
