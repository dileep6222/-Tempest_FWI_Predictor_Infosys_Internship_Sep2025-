```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv(r'E:\Info_Srping\FWI_Prediction\Data\raw\raw_dataset1.csv')
dataset
```

```python
dataset.info(verbose=True, show_counts=True)
```

```python
dataset.drop(['aes','wmo','calcstatus','tz','opts'], axis=1, inplace=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
column_mapping = {
    'rep_date': 'report_date',
    'temp': 'temperature',
    'td': 'dew_point_temperature',
    'rh': 'relative_humidity',
    'ws': 'wind_speed',
    'wg': 'wind_gust',
    'wdir': 'wind_direction',
    'pres': 'atmospheric_pressure',
    'vis': 'visibility',
    'precip': 'precipitation',
    'rndays': 'rain_days',
    'sog': 'snow_on_ground',
    'ffmc': 'fine_fuel_moisture_code',
    'dmc': 'duff_moisture_code',
    'dc': 'drought_code',
    'bui': 'build_up_index',
    'isi': 'initial_spread_index',
    'fwi': 'fire_weather_index',
    'dsr': 'daily_severity_rating',
    'country': 'country',
    'lat': 'latitude',
    'lon': 'longitude',
    'elev': 'elevation'
}
```

```python
dataset.rename(columns=column_mapping, inplace=True)
```

```python
dataset
```

```python
dataset.drop(['wind_gust'], axis = 1, inplace = True)
dataset
```

```python
print("Before optimization:")
dataset.info(memory_usage='deep')
```

```python
import numpy as np

# Convert float64 → float32 / float16
for col in dataset.select_dtypes(include=['float64']).columns:
    dataset[col] = pd.to_numeric(dataset[col], downcast='float')

# Convert int64 → int32 / int16
for col in dataset.select_dtypes(include=['int64']).columns:
    dataset[col] = pd.to_numeric(dataset[col], downcast='integer')
```

```python
dataset
```

```python
print("After optimization:")
dataset.info(memory_usage='deep')
```

```python
dataset.info(verbose=True, show_counts=True)
```

```python
dataset.dropna(inplace = True)
dataset
```

```python
dataset.info(verbose=True, show_counts=True)
```

```python
dataset.to_csv(r'E:\Info_Srping\FWI_Prediction\Data\raw\rae_dataset_v1')
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_data, temp_data = train_test_split(
     dataset,
    test_size=0.4,
    stratify=dataset['country'],
    random_state=42
)
```

```python
train_data
```

```python
temp_data
```

```python
dev_data, test_data = train_test_split(
     temp_data,
     test_size=0.4,
     stratify=temp_data['country'],
     random_state=42
)
```

```python
print("Train size:", len(train_data))
print("Dev size:", len(dev_data))
print("Test size:", len(test_data))

print("Country distribution in original:")
print(dataset['country'].value_counts(normalize=True))

print("Country distribution in Train:")
print(train_data['country'].value_counts(normalize=True))

print("Country distribution in Dev:")
print(dev_data['country'].value_counts(normalize=True))

print("Country distribution in Test:")
print(test_data['country'].value_counts(normalize=True))
```

```python
train_data.to_csv(r'E:\Info_Srping\FWI_Prediction\Data\processed\train_data.csv', index = False)
dev_data.to_csv(r'E:\Info_Srping\FWI_Prediction\Data\processed\dev_data.csv', index = False)
test_data.to_csv(r'E:\Info_Srping\FWI_Prediction\Data\processed\test_data.csv', index = False)
```
